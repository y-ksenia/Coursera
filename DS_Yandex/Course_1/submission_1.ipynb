```python
import numpy as np
import scipy as sp
import re
import pandas as pd
```

```python
data = np.array([i.lower().strip() for i in open('sentences.txt')])
```

```python
new_data = []
for i in range(len(data)):
    new_data.append(np.array(re.split('[^a-z]',data[i])))
    new_data[i] = new_data[i][new_data[i] != '']
```

```python
new_data[9]
```

```python
s = set()
for i in range(len(new_data)):
    s = s.union(set(new_data[i]))
```

```python
dic = dict()
dic = dic.fromkeys(s)

for i in range(len(new_data)):
    for j in range(len(new_data[i])):
        for key in dic.keys():
            if key == new_data[i][j]:
                if dic[key] == None:
                    dic[key] = [(i,j)]
                else:
                    dic[key].append((i,j))
dic
```

```python
new_dic = dict()
new_dic = new_dic.fromkeys(s)

from collections import Counter
Counter.keys
for key in dic.keys():
    dic[key] = np.array(dic[key])
    new_dic[key] = np.array(list(Counter(dic[key][:,0]).items()))
# print(Counter.keys())
new_dic
```

```python
df = pd.DataFrame(index = s, columns = range(len(new_data)))
```

```python
df = df.sort_index(axis=0)
```

```python
for i in df.index:
    for j in range(len(new_dic[i])):
        df.set_value(i, new_dic[i][j,0], new_dic[i][j,1])
```

```python
df = df.fillna(0)
```

```python
from scipy.spatial.distance import cosine 
dist = []
min_dist = 0
min_i = 0
for i in df.columns:
    dist.append(cosine(df[0],df[i]))
    if min_dist > cosine(df[0],df[i]):
        min_dist = cosine(df[0],df[i])
        min_i = i
```
