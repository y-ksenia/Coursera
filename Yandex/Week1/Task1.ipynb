```python
from collections import defaultdict

girl_names = df['Name'][df['Sex'] == 'female'].values
count_names = defaultdict(int)
for name in girl_names:
    for word in name.split('.')[1].replace('(','').replace(')','').split():
        count_names[word] += 1
```

```python
import operator

sorted_d = sorted(count_names.items(), key=operator.itemgetter(1))
```
